```python
from pysdkit.data import test_multivariate_signal
from pysdkit.plot import plot_signal, plot_IMFs

time, signal = test_multivariate_signal(case=4)
fig = plot_signal(time, signal)
```

```python
from pysdkit import MVMD

mvmd = MVMD(K=2, alpha=2000, tau=0.0)
results = mvmd.fit_transform(signal)
```

```python
print(results.shape)
```

```python
fig = plot_IMFs(signal, results, return_figure=True)
fig.savefig("MVMD_decomposition_results_2D.jpg", bbox_inches="tight", 
            dpi=800)
```

```python
fig = plot_IMFs(signal, results, view="3d", return_figure=True)
fig.savefig("MVMD_decomposition_results_3D.jpg", bbox_inches="tight",
            dpi=800)
```
