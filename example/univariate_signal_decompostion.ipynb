```python
from pysdkit.data import test_univariate_signal
from pysdkit.plot import plot_IMFs
from pysdkit import EMD

# 创建用于测试的实例
time, signal = test_univariate_signal(case=1)

# 创建EMD算法的信号分解对象并传入参数
emd = EMD(max_imfs=3)

# 执行信号分解算法
results = emd.fit_transform(signal)
# 或 results = vmd(signal)

# 打印信号的尺度大小
print(results.shape)

# 对分解的结果进行可视化
fig = plot_IMFs(signal, results, return_figure=True, view="3D")
```

```python
from pysdkit.plot import plot_IMFs_amplitude_spectra

# frequency domain visualization
plot_IMFs_amplitude_spectra(results, smooth="exp")   # use exp smooth
```

```python
fig.savefig("emd_decomposition_3D.jpg", bbox_inches="tight", dpi=1000)
```

```python
from pysdkit.data import test_univariate_signal
from pysdkit.plot import plot_IMFs
from pysdkit import VMD

# 创建用于测试的实例
time, signal = test_univariate_signal(case=1)

# 创建VMD算法的信号分解对象并传入参数
vmd = VMD(alpha=800,
          K=3,  # 待分解模态的数目
          tau=0.0)

# 执行信号分解算法
results = vmd.fit_transform(signal)
# 或 results = vmd(signal)

# 打印信号的尺度大小
print(results.shape)

# 对分解的结果进行可视化
plot_IMFs(signal, results)
```

```python
from pysdkit.plot import plot_IMFs_amplitude_spectra

# frequency domain visualization
plot_IMFs_amplitude_spectra(results, figsize=(10, 5), smooth="exp", save_name="fft", save_figure=True)   # use exp smooth
```
