```python
from pysdkit import VMD2D
from pysdkit.data import test_grayscale
from pysdkit.plot import plot_grayscale_image
```

```python
# 创建用于测试的图像数据
image = test_grayscale()
```

```python
# 打印并检验图像的尺寸
print(image.shape)
```

```python
plot_grayscale_image(image)
```

```python
# 实例化用于二维图像的VMD算法
vmd2d = VMD2D(
    K=5, alpha=5000, tau=0.25, DC=True, init="random", tol=1e-6, max_iter=3000
)
u = vmd2d.fit_transform(image)
print(IMFs.shape)
```

```python
from matplotlib import pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']  #显示中文
plt.rcParams['axes.unicode_minus'] = False  #显示负号

from pysdkit.utils import fft2d, fftshift

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 7.2),
                       sharex=True, sharey=True, )
fig.subplots_adjust(wspace=0.05, hspace=0.10)

ax[0][0].imshow(np.abs(fft2d(image)), cmap="coolwarm")
ax[0][0].set_title("原始输入图像", fontsize=14)

g = (i for i in range(1, 10))

ax[0][1].imshow(np.abs(fft2d(u[:, :, 0])), cmap="coolwarm")
ax[0][1].set_title(f"分解模态{next(g)}", fontsize=14)
ax[0][2].imshow(np.abs(fft2d(u[:, :, 1])), cmap="coolwarm")
ax[0][2].set_title(f"分解模态{next(g)}", fontsize=14)
ax[1][0].imshow(np.abs(fft2d(u[:, :, 2])), cmap="coolwarm")
ax[1][0].set_title(f"分解模态{next(g)}", fontsize=14)
ax[1][1].imshow(np.abs(fft2d(u[:, :, 3])), cmap="coolwarm")
ax[1][1].set_title(f"分解模态{next(g)}", fontsize=14)
ax[1][2].imshow(np.abs(fft2d(u[:, :, 4])), cmap="coolwarm")
ax[1][2].set_title(f"分解模态{next(g)}", fontsize=14)
```
